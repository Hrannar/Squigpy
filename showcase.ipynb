# Squigpy
This notebook demonstrates the features of Squigpy - a code implementation of the squigonometric functions in Python. Squigpy was developed for a master thesis project in Astrophysics at Lund University by Viktor Hrannar JÃ³nsson.

In short, squigonometry is a generalization of trigonometry defined by the unit squircle isntead of the unit circle. See Chebolu et al. (2021) for a more detailed explanation of the mathematics. 

```python
from squigpy import pi_p, squine, cosquine, tanquent
import matplotlib.pyplot as plt
import numpy as np
```

## $\pi_p$
The circumference of any unit squircle is not $2\pi$, except in the case of the unit circle where $p=2$. Instead, the circumference is described by the generalized constant $\pi_p$ for any given $p$.

As $p\rightarrow\infty$, $\pi_p\rightarrow 4$ because the unit squircle at that p-value is a square with sidelength 2. Conversely, at $p=1$, the unti squircle is a square with sidelength 1 and hence $\pi_p=2$.

The squigpy function pi_p takes *array_like* $p$-values and returns the corresponding $\pi_p$-values.

```python
# Single p-values as arguments
print(r'p = 1 : pi_p =', pi_p(1))
print(r'p = 2 : pi_p =', pi_p(2))
print(r'p = 5 : pi_p =', pi_p(5))
# p-values in array as arguments
p_array = np.array([1,2,5])
print(r'p = ', p_array, r'  pi_p = ', pi_p(p_array))
```

```python
p_array = np.linspace(1,50, 1000)   # Array of p-values in range (1,50)

plt.figure(figsize=(6,3))
plt.plot(p_array, pi_p(p_array))   # Plot pi_p for each value in p_array
plt.xscale('log')
plt.grid(alpha=0.6)
plt.ylim(2,4.05)
plt.xlim(1,50)

plt.plot([2, 2], [2,np.pi], '--k', alpha=0.6)
plt.plot([1, 2], [np.pi,np.pi], '--k', alpha=0.6)
plt.text(2.1, 3.1, r'$p=2:\pi_p=\pi$', alpha=0.8)

plt.xlabel(r'$p$', fontsize=14)
plt.ylabel(r'$\pi_p$', fontsize=14)
plt.title(r'$\pi_p: p\rightarrow \infty$', fontsize=16)

plt.show()
```

## Squine and Cosquine
The Squigpy functions squine and cosquine takes two arguments. The first is the angle values $\theta$ (*array_like*) for which you wish to evaluate the squigonometric function at, and the second argument is the $p$-value (*int* or *float*).

Note that it is not possible to insert angle values larger $\theta<0$ or $\theta>2\pi_p$.

```python
p=3
print(squine(0, p))
print(squine([0,1], p))
print(squine(np.array([0,1,2]), p))
```

```python
fig, ax = plt.subplots(3, 1, sharex=True)

p = 2
t = np.linspace(0,2*pi_p(p))
ax[0].plot(t, squine(t, p), label='squine')
ax[0].plot(t, cosquine(t, p), label='cosquine')
ax[0].text(3.5, 0.7, r'$p=2$')

p = 5
t = np.linspace(0,2*pi_p(p))
ax[1].plot(t, squine(t, p))
ax[1].plot(t, cosquine(t, p))
ax[1].text(3.5, 0.7, r'$p=5$')

p = 100
t = np.linspace(0,2*pi_p(p))
ax[2].plot(t, squine(t, p))
ax[2].plot(t, cosquine(t, p))
ax[2].text(3.5, 0.7, r'$p=100$')


ax[2].set_xlabel(r'$\theta$', fontsize=14)
ax[0].legend()

plt.show()
```

## Tanquent function
Like the squine and cosquine functions, but the generalization of the tangent.

```python
p=3
print(tanquent(0, p))
print(tanquent([0,1], p))
print(tanquent(np.array([0,1,2]), p))
```

```python
p = 2
t = np.linspace(0,2*pi_p(p), 1000)
plt.plot(t, tanquent(t, p), label='p=2')
plt.ylim(-5,5)

p = 5
t = np.linspace(0,2*pi_p(p), 1000)
plt.plot(t, tanquent(t, p), label='p=5')
plt.ylim(-5,5)

p = 100
t = np.linspace(0,2*pi_p(p), 1000)
plt.plot(t, tanquent(t, p), label='p=100')
plt.ylim(-5,5)

plt.grid(alpha=0.6)
plt.legend()
plt.xlabel(r'$\theta$', fontsize=14)
plt.title('Tanquent function', fontsize=16)

plt.show()
```
